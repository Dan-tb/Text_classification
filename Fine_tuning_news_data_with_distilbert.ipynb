```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, DataCollatorWithPadding, TrainingArguments, Trainer
import numpy as np
from torch.utils.data import DataLoader
from datasets import Dataset, DatasetDict
import pandas as pd
```

```python
news_category = pd.read_json('train/News_Category_Dataset_v3.json', lines=True)
news_category.drop(['authors', 'link', 'date'], axis = 1, inplace = True)
news_category.head()
```

## when running on local machine.

```python
news_category = pd.read_json('News_Category_Dataset_v3.json', lines=True)
news_category.drop(['authors', 'link', 'date'], axis = 1, inplace = True)
news_category.head()
```

## Applies anywhere

```python
checkpoint_1 ='distilbert-base-uncased'
tokenizer = AutoTokenizer.from_pretrained(checkpoint_1)
model = AutoModelForSequenceClassification.from_pretrained(checkpoint_1, num_labels=42)
```

```python
news_category.info()
```

```python
len(news_category['category'].value_counts().index)
```

```python
# Split the data into train and evaluation (stratified)
from sklearn.model_selection import train_test_split
df_train, df_eval = train_test_split(news_category, train_size=0.7,stratify=news_category.category, random_state=42) # Stratified splitting
```

```python
raw_datasets = DatasetDict({
    "train": Dataset.from_pandas(df_train),
    "eval": Dataset.from_pandas(df_eval)
})
```

```python
raw_datasets = raw_datasets.class_encode_column('category')
```

```python
raw_datasets['train'][7]
```

```python
tokenized_datasets = raw_datasets.map(lambda news_category: tokenizer(news_category['headline'], news_category['short_description'], truncation=True), batched=True)
print(tokenized_datasets)
```

```python
print(tokenized_datasets["train"][0])
```

```python
tokenized_datasets = tokenized_datasets.remove_columns(["headline", "__index_level_0__", "short_description"])
tokenized_datasets = tokenized_datasets.rename_column("category", "labels")
tokenized_datasets.set_format("torch")
print(tokenized_datasets)
```

```python
tokenized_datasets['train'][0]
```

```python
# Padding for batch of data that will be fed into model for training
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
train_dataloader = DataLoader(
    tokenized_datasets["train"], shuffle=True, batch_size=8, collate_fn=data_collator
)
eval_dataloader = DataLoader(
    tokenized_datasets["eval"], batch_size=8, collate_fn=data_collator
)
```

```python
for batch in train_dataloader:
    break
{k: v.shape for k, v in batch.items()}
```

```python
training_args = TrainingArguments("test-trainer")
```

```python
trainer = Trainer(
    model,
    training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["eval"],
    data_collator=data_collator,
    tokenizer=tokenizer,
)
```

```python
trainer.train()
```

```python
from sklearn.metrics import classification_report

# Make prediction on evaluation dataset
y_pred = trainer.predict(tokenized_datasets["eval"]).predictions
y_pred = np.argmax(y_pred, axis=-1)

# Get the true labels
y_true = tokenized_datasets["eval"]["labels"]
y_true = np.array(y_true)

# Print the classification report
print(classification_report(y_true, y_pred, digits=3))
```
